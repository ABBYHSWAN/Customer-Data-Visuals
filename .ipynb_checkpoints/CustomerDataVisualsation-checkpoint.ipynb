```python
data = pd.concat(map(pd.read_csv, ["customers_export_1.csv", "customers_export_2.csv", 
     "customers_export_3.csv","customers_export_4.csv"]))

#print(data.shape)
#data.info(memory_usage='deep')
data.sample(5)
```

```python
def compress(data, verbose=True):
    numerics = ['int64', 'float64', 'object']
    start_mem = data.memory_usage().sum() / 1024**2    
    for col in data.columns:
        col_type = data[col].dtypes
        if col_type in numerics:
            c_min = data[col].min()
            c_max = data[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    data[col] = data[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    data[col] = data[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    data[col] = data[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    data[col] = data[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    data[col] = data[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    data[col] = data[col].astype(np.float32)
                else:
                    data[col] = data[col].astype(np.float64)    
    end_mem = data.memory_usage().sum() / 1024**2
    if verbose: 
        print('Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction)'.format(end_mem, 100 * 
                                                                              (start_mem - end_mem) / start_mem))
    return data

train_df = compress(train_df)
train_df.info(memory_usage='deep')
```

```python
province_dict = dict()

for row in data.dropna(subset = 'Province')['Province']:
    if row not in province_dict:
        province_dict[row] = 1
    else:
        province_dict[row] = province_dict[row] + 1
```

```python
cut_prov_dict = province_dict
count = 0
keys_list = list(cut_prov_dict)

while count < len(keys_list):
    key = keys_list[count]
    if cut_prov_dict[key] < 100: #minimum customer count per province 
        cut_prov_dict.pop(key)
    count += 1
    
cut_prov_dict = dict(sorted(cut_prov_dict.items(), key=lambda x: x[1], reverse = True))
```

Create a list and corresponding pie chart that shows the distribution of customers's total spending

```python
spent_list = data['Total Spent'].tolist()
spent_dict = {'< $25':0, '< $50':0, '< $75':0, '< $100':0, '< $500':0, '> $500':0}

for total in spent_list:
    if total != 0.0:
        if total < 25.0:
            spent_dict['< $25'] = spent_dict['< $25'] + 1
        elif total < 50.0:
            spent_dict['< $50'] = spent_dict['< $50'] + 1
        elif total < 75.0:
            spent_dict['< $75'] = spent_dict['< $75'] + 1
        elif total < 100.0:
            spent_dict['< $100'] = spent_dict['< $100'] + 1
        elif total < 500.0:
            spent_dict['< $500'] = spent_dict['< $500'] + 1
        elif total > 500.0:
            spent_dict['> $500'] = spent_dict['> $500'] + 1
```

```python
total_spent = list(spent_dict.keys())
customer_count = list(spent_dict.values())


colors = ['#f7e4ee','#ffb3fc','#fcebfc', 'pink','#ddace3', '#f797c6']
plt.figure(figsize = (8, 8))
plt.pie(customer_count, labels= total_spent, colors = colors)
plt.title("Totals Spent by Customers")
plt.show()
```

```python
orders_dict = dict()

for row in data['Total Orders']:
    if row != 0:
        if row not in orders_dict:
            orders_dict[row] = 1
        else:
            orders_dict[row] = orders_dict[row] + 1
            
orders_dict = collections.OrderedDict(sorted(orders_dict.items()))
```

```python
cut_dict = orders_dict
count = 0
keys_list = list(cut_dict)

while count < len(keys_list):
    key = keys_list[count]
    if cut_dict[key] < 100: #minimum frequncy of customers who have placed a specific number of orders
        cut_dict.pop(key)
    count += 1
```
